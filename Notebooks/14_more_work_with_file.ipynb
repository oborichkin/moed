```python
import sys
from random import random
sys.path.append('../')
from moed.model import Sequence, Model
from moed.analysis import Analysis
from moed.proc import Proc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import struct 

X = [x/1000 for x in range(1000)]
with open("../data/pgp_f4-1K-1ms.dat", "br") as f:
    Y = list(struct.unpack('1000f', f.read()))
```

```python
from collections import OrderedDict

signal = Sequence.from_dict(OrderedDict(zip(X,Y)))
signal.plot()
```

#    

```python
ac = Analysis.auto_correlation(signal)
```

```python
ac.plot()
```

```python
Analysis.dft(signal).plot()
reversed_twice = Analysis.idft(Analysis.dft_complex(signal))
reversed_twice.plot()
```

```python
cc = Analysis.cross_correlation(signal, reversed_twice)
```
